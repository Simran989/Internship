```python
#2. Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release)
#and make data frame.?
```

```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup 
```

```python
url = 'https://www.imdb.com/search/title/?count=100&groups=top_1000&sort=user_rating'
```

```python
page = requests.get(url)
page
```

```python
soup = BeautifulSoup(page.content, "html.parser")
```

```python
#Making empty list for append
movie_name=[]
Rate=[]
year=[]
```

```python
#Sorting data
movie_data = soup.find_all('div', class_ = 'lister-item mode-advanced')
```

```python
for store in movie_data:
    name = store.h3.a.text
    movie_name.append(name)
    
    year_of_release = store.h3.find('span', class_= 'lister-item-year text-muted unbold').text.replace('\n',' ').replace(')','')
    year.append(year_of_release)
    
    Rating = store.find('div', class_= 'inline-block ratings-imdb-rating').text.replace('\n', '')
    Rate.append(Rating)
```

```python
data = pd.DataFrame({'Movie Name':movie_name, 'Year of release':year, 'Rating': Rate})
```

```python
#Ques 3 Write a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. name, rating, year of
#release) and make data frame.
```

```python
url = 'https://www.imdb.com/india/top-rated-indian-movies/'
```

```python
page = requests.get(url)
page
```

```python
soup = BeautifulSoup(page.content, "html.parser")
```

```python
scraped_movies = soup.find_all('td', class_= 'titleColumn')
```

```python
movies = []
for movie in scraped_movies:
    movie = movie.get_text().replace('\n', "")
    movie = movie.strip(" ")
    movies.append(movie)
movies
```

```python
#get rating for movies
scraped_ratings= soup.find_all('td', class_= 'ratingColumn imdbRating')
scraped_ratings
```

```python
#parse rating 
ratings = []
for rating in scraped_ratings:
    rating = rating.get_text().replace('\n', '')
    ratings.append(rating)
ratings
```

```python
data = pd.DataFrame()
data['Movie Names and year']= movies
data['Ratings'] = ratings
```
